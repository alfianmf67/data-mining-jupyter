```python
dataset = pd.read_csv("Data Intelligent Irrigation System.csv")
```

```python
x = dataset.iloc[:, [1,2]].values #memuat data
y = dataset.iloc[:, -1].values #label
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

SCalling data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

CALL Function KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(x_train, y_train)
```

Menentukan prediksi

```python
y_pred = classifier.predict(x_test)
```

evaluasi data dan melakukan validasi

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
akurasi = classification_report(y_test,y_pred)
```

```python
print(akurasi)
```

```python
from sklearn.metrics import accuracy_score
akurasi = accuracy_score(y_test,y_pred) 
print("Tingkat Akurasi :%d persen"%(akurasi*100))
```

membuat visualisasi data

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min()-1, stop = x_set[:, 0].max()+1, step = 0.01),
                     np.arange(start = x_set[:, 1].min()-1, stop = x_set[:, 0].max()+1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape), alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate (np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan K-Nearest Neighbor')
plt.xlabel('Moisture')
plt.ylabel('Temp')
plt.legend()
plt.show()
```
